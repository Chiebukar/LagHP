```python
def for_price(x):
    a = x.split(',')
    b = ''
    for i in range(len(a)):
        b += a[i]
    return b
```

```python
for_price('2,480,000')
```

```python
la4['Price'] = la4['Price'].apply(for_price)
```

```python
la4['Price'].unique()
```

```python
def for_location(x):
    a = x.split()
    return a[-2]
```
