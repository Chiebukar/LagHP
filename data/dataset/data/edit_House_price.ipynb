```python
def splitx(x):
    a = str(x)
    b = x.split(',')
    if len(b) == 2:
        val = b[0] + b[1]
        return val
    if len(b) == 3:
        val = b[0] + b[1] + b[2]
        return val   
    else:
        return b
    return b
```

```python
splitx('6,000,000')
```

```python
la1['Price'] = la1['Price'].apply(splitx)
la1.head()
```

```python
def locate(x):
    a = str(x)
    b = a.split(',')
    index = b.index(' Lagos')
    return b[index-1]
```
