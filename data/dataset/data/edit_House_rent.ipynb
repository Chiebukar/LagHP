```python
def locate(x):
    a = str(x)
    b = a.split(',')
    if ' Lagos' in b:
        index = b.index(' Lagos')
        return b[index-1]
    else:
        return None
```

```python
def format_price(x):
    a =  x.split()
    b = a[1]. split(',')
    if len(b) == 2:
        val = b[0] + b[1]
        return val
    elif len(b) == 3:
        val = b[0] + b[1] + b[2]
        return val
    else:
        return None
```

```python
def format_beds(x):
    x = str(x)
    b = x.split()
    if 'm²' in b:
        return None
    return b[0]
```
